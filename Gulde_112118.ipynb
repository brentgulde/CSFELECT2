```
import matplotlib.pyplot as plt
import numpy as np
a = [1,2,3,4]
b = [5,6,7,8]
c = np.multiply(a,b)
print(a)
```

```
print(b)
```

```
plt.plot(a,a,"r")
plt.ylabel("values of a")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```
plt.plot(a,b,"g")
plt.ylabel("values of b")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```

plt.plot(a,c,"b")
plt.ylabel("values of c")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```
plt.plot(a,a,"r",a,b,"g",a,c,"b")
plt.ylabel("values of a, b and c")
plt.xlabel("values of a")
plt.title("Graph with three (3) Y axes")
plt.show()
```

```
plt.figure(1)

plt.subplot(131)
plt.plot(a,a,"r")
plt.ylabel("values of a")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.subplot(132)
plt.plot(a,b,"g")
plt.ylabel("values of b")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.subplot(133)
plt.plot(a,c,"b")
plt.ylabel("values of c")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.title("Grap in Equal Size of Axes")
plt.show
```

```
d = [5,10,15,30]
```

```
plt.plot(a,a,"r",a,b,"g",a,c,"b", a,d,"g")
plt.ylabel("values of a, b, c and d")
plt.xlabel("values of a")
plt.title("Graph with four (4) Y axes")
plt.show()
```

```
plt.figure(1)

plt.subplot(131)
plt.plot(a,a,"r")
plt.ylabel("values of a")
plt.xlabel("values of a")
plt.ylim(0,max(d))

plt.subplot(132)
plt.plot(a,b,"g")
plt.ylabel("values of b")
plt.xlabel("values of a")
plt.ylim(0,max(d))

plt.subplot(133)
plt.plot(a,c,"b", a,d,"g")
plt.ylabel("values of c and d")
plt.xlabel("values of a")
plt.ylim(0,max(d))



plt.title("Grap in Equal Size of Axes")
plt.show
```
